```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
def draw_confirmed_cases_lin(country, province=''):

    confirmed = np.array(covid19_confirmed[(covid19_confirmed['Country/Region']==country) 
                                         & (covid19_confirmed['Province/State']==province)].iloc[0,4:])
    plt.plot(confirmed, label = country + ' ' + province)
    
def draw_confirmed_cases_log(country, province=''):

    confirmed = np.array(covid19_confirmed[(covid19_confirmed['Country/Region']==country) 
                                         & (covid19_confirmed['Province/State']==province)].iloc[0,4:])
    
    plt.semilogy(confirmed, label = country + ' ' + province)

def draw_deaths_lin(country, province=''):

    confirmed = np.array(covid19_deaths[(covid19_deaths['Country/Region']==country) 
                                      & (covid19_deaths['Province/State']==province)].iloc[0,4:])
    plt.plot(confirmed, label = country + ' ' + province)
    
def draw_deaths_log(country, province=''):

    confirmed = np.array(covid19_deaths[(covid19_deaths['Country/Region']==country) 
                                         & (covid19_deaths['Province/State']==province)].iloc[0,4:])
    
    plt.semilogy(confirmed, label = country + ' ' + province)    
```

```python
url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/"
covid19_confirmed = pd.read_csv(url + "time_series_covid19_confirmed_global.csv")
covid19_confirmed['Province/State'].fillna('', inplace=True)
covid19_deaths = pd.read_csv(url + "time_series_covid19_deaths_global.csv")
covid19_deaths[   'Province/State'].fillna('', inplace=True)
covid19_recovered = pd.read_csv(url + "time_series_covid19_recovered_global.csv")
covid19_recovered['Province/State'].fillna('', inplace=True)
import datetime
dates = covid19_confirmed.columns[4:]
dates = np.array([datetime.datetime.strptime(x, '%m/%d/%y').strftime('%d/%m') for x in dates])
T = len(dates)
t = np.arange(T)
```

```python
plt.figure(figsize = (8, 6))
plt.title('Confirmed cases (logarithmic)')
draw_confirmed_cases_log('Russia')
draw_confirmed_cases_log('France')
draw_confirmed_cases_log('China', 'Hubei')
draw_confirmed_cases_log('US')
draw_confirmed_cases_log('Italy')
draw_confirmed_cases_log('Korea, South')
draw_confirmed_cases_log('Germany')
draw_confirmed_cases_log('Japan')
draw_confirmed_cases_log('Spain')
draw_confirmed_cases_log('Israel')
plt.grid()
plt.legend()
plt.xticks(t[0::10], dates[0::10])
pass
```

```python
plt.figure(figsize = (9, 6))

plt.title('Deaths (logarithmic)')
draw_deaths_log('Russia')
draw_deaths_log('France')
draw_deaths_log('China', 'Hubei')
draw_deaths_log('US')
draw_deaths_log('Italy')
draw_deaths_log('Korea, South')
draw_deaths_log('Germany')
draw_deaths_log('Japan')
draw_deaths_log('Spain')
draw_deaths_log('Israel')
plt.grid()
plt.legend()
plt.xticks(t[0::10], dates[0::10])
pass
```
